```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
```

```python
(X_train,Y_train),(X_test,Y_test)=fashion_mnist.load_data()
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
plt.figure(figsize=(10, 10))
for i in range(15):
    plt.subplot(5, 5, i + 1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(class_names[Y_train[i]])
    plt.axis('off')
plt.show()
```

```python
#  Data Preprocesiing
```

```python
X_train=X_train.astype('float32')
X_test=X_test.astype('float32')
```

```python
RESHAPED=784
```

```python
X_train=X_train.reshape(60000,RESHAPED)
```

```python
X_test=X_test.reshape(10000,RESHAPED)
```

```python
X_train/=255
X_test/=255
```

```python
print(X_train.shape,'train sample')
print(X_test.shape,'test samples')
```

```python
Y_train=tf.keras.utils.to_categorical(Y_train,10)
Y_test=tf.keras.utils.to_categorical(Y_test,10)
```

```python
# Single layer Perceptron
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
model_0 = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')  # Output layer with 10 units for 10 classes
])
```

```python
model_0.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
training=model_0.fit(X_train,Y_train,batch_size=64,epochs=70,validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
print(training.history.keys())
```

```python
plt.plot(training.history['accuracy'])
plt.plot(training.history['val_accuracy'])
plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(training.history['loss'])
plt.plot(training.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
test_loss,test_acc=model_0.evaluate(X_test,Y_test)
print('Test accuracy:',test_acc)
```

```python
model_0.summary()
```

```python
from tensorflow.keras.layers import Dropout
#Most common type of model is a stack of layers
model_2 = tf.keras. Sequential()
N_hidden = 128
# Adds a densely-connected layer with 64 units to the model:
model_2.add(Dense(N_hidden, name='dense_layer', input_shape=(784,), activation = 'relu'))
model_2.add(Dropout (0.3))
# Adding another dense layer:
model_2.add(Dense(N_hidden, name='dense_layer_2', activation='relu'))
model_2.add(Dropout(0.3))
# Add an output layer with 10 output units (10 different classes):
model_2.add(Dense(10, name='dense_layer_3', activation = 'softmax'))
```

```python
model_2.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
training=model_2.fit(X_train,Y_train,batch_size=64,epochs=50,validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
print(training.history.keys())
```

```python
plt.plot(training.history['accuracy'])
plt.plot(training.history['val_accuracy'])
plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(training.history['loss'])
plt.plot(training.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
model_2.weights[0]
```

```python
test_loss,test_acc=model_2.evaluate(X_test,Y_test)
print('Test accuracy:',test_acc)
```
